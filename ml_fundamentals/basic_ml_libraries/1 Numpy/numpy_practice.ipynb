# Numpy Practice

## 1. Import Numpy

```
import numpy as np
```

## 2. Create a 1D array of numbers from 0 to 9

```
arr1 = np.arange(10)
print(arr1)
```

## 3. Create a 3x3 numpy array of all True's

```
arr2 = np.ones((3, 3), dtype=bool)
print(arr2)
```

## 4. Extract all odd numbers from arr1

```
arr_odd = arr1[arr1 % 2 == 1]
print(arr_odd)
```

## 5. Replace all odd numbers in arr1 with -1

```
arr1[arr1 % 2 == 1] = -1
print(arr1)
```
