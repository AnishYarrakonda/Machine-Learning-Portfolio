# Pandas Practice

## 1. Import pandas

```
import pandas as pd
import numpy as np
```

## 2. Create a series from a list, numpy array and dict

```
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser1 = pd.Series(mylist)
ser2 = pd.Series(myarr)
ser3 = pd.Series(mydict)
print(ser3.head())
```

## 3. Convert the series ser3 into a dataframe with its index as another column

```
df = ser3.to_frame().reset_index()
df.columns = ['letter', 'number']
print(df.head())
```

## 4. Combine ser1 and ser2 to form a dataframe

```
df2 = pd.concat([ser1, ser2], axis=1)
print(df2.head())
```
