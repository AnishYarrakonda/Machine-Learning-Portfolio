# Matplotlib Practice

## 1. Import pyplot

```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 2. Create a simple plot of x vs y

```
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 3. Create a scatter plot

```
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, alpha=0.5)
plt.title('Random Scatter Plot')
plt.show()
```
